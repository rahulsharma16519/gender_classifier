```
import numpy as np
from keras.preprocessing import image
import pickle
```

```
np.set_printoptions(precision=3, suppress=False)
```

```
test_image = image.load_img('/content/drive/My Drive/IMG_20170222_231203.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
```

```
classifier = pickle.load(open('/content/drive/My Drive/genderClassifier.pckl', 'rb'))
```

```
result = classifier.predict_classes(test_image)

if result[0][0] == 1:
    prediction = 'woman'
else:
    prediction = 'man'
```
